```python
#Categorical column distribution
def Category_distrubution(df):
    for i in df.columns:
        if df[i].dtype=='O':
            sns.barplot(df[i])
            plt.title(i)
            plt.show()
    
```

```python
#Numerical column distribution
def Numerical_distrubution(df):
    for i in df.columns:
        if df[i].dtype in ['int64','float64']:
            plt.hist(df[i])
            plt.title(i)
            plt.show()
```

Now checking if the distribution is follows normal distribution or not

```python
def normal_distribution(df):
    for i in df.columns:
        if df[i].dtype!='O':
            sns.kdeplot(df[i])
            plt.title(i)
            plt.show()
```

```python
normal_distribution(df)
```

```python
#Now assigning the target to the y variable and features to the X Variable 
x=df.drop(columns=['High_Default_Risk'],errors='ignore')
y=df['High_Default_Risk']
```

```python
#Split the data into train and test
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
x_train.info()
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OrdinalEncoder,RobustScaler
from sklearn.pipeline import Pipeline
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
from sklearn.neighbors import KNeighborsClassifier
transformer=ColumnTransformer([("t1",RobustScaler(),[2,4,13,15]),("t2",OrdinalEncoder(),[0,1,6,10,12])],remainder='passthrough')
```

```python
#KNN Classifer
from sklearn.neighbors import KNeighborsClassifier
knn_pipe = Pipeline([("transformer", transformer),("model", KNeighborsClassifier())])
#Fit the data
knn_pipe.fit(x_train, y_train)
y_pred =knn_pipe.predict(x_test)
print("Accuracy:", accuracy_score(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

```python
#Decision Tree Classifier
from sklearn.neighbors import DecisionTreeClassifier
dt_pipe = Pipeline([("transformer", transformer),("model", DecisionTreeClassifier())])
#Fit the data
dt_pipe.fit(x_train, y_train)
y_pred =dt_pipe.predict(x_test)
print("Accuracy:", accuracy_score(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

```python
from sklearn.tree import LogisticRegression
lr_pipe = Pipeline([("transformer", transformer),("model", LogisticRegression())])
#Fit the data
lr_pipe.fit(x_train, y_train)
y_pred =lr_pipe.predict(x_test)
print("Accuracy:", accuracy_score(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```
